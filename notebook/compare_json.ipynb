```python
import json
import pandas as pd
import json

# CSV 파일 경로
csv_file_path = 'gold.csv'

# CSV 파일을 DataFrame으로 읽기
df = pd.read_csv(csv_file_path)

# DataFrame을 JSON 형식의 문자열로 변환
json_str = df.to_json(orient='records')

# JSON 문자열을 Python 딕셔너리로 변환
data = json.loads(json_str)

data1 = {item['id']: item['answers'] for item in data}
print(data1)

# JSON 데이터 불러오기
#with open('augmented_train_baseline+BM25_roberta_large_b16.json', 'r', encoding='utf-8') as f:
 #   data1 = json.load(f)
    
with open('augmented_train_baseline+BM25_roberta_large_b16.json', 'r', encoding='utf-8') as f:
    data2 = json.load(f)
print(data2)
# 공통 키 찾기
common_keys = set(data1.keys()) & set(data2.keys())

# 값이 다른 공통 키 찾기
different_values = {key: (data1[key], data2[key]) for key in common_keys if data1[key] != data2[key]}

# 값이 다르지만 서로 포함하는 공통 키 찾기
#different_but_contain_values = {key: (data1[key], data2[key]) for key in common_keys if (data1[key] != data2[key]) & ((data1[key] in data2[key]) | (data2[key] in data1[key]))}
different_but_contain_values = {
    key: (data1[key], data2[key]) 
    for key in common_keys 
    if data1[key] is not None and data2[key] is not None # NoneType 체크 추가
    and isinstance(data1[key], str) and isinstance(data2[key], str) # 문자열 타입 체크 추가
    and data1[key] != data2[key] 
    and ((data1[key] in data2[key]) or (data2[key] in data1[key]))
}

# 첫 번째 데이터에만 있는 값 찾기
unique_to_data1 = set(data1.values()) - set(data2.values())

# 두 번째 데이터에만 있는 값 찾기
unique_to_data2 = set(data2.values()) - set(data1.values())

# 결과 출력
print(f"공통 키 중에서 값이 다른 키와 해당 값들: {different_values}")
print(f"일치율: {(len(data1)-len(different_values))/len(data1)*100}")
print(f"공통 키 중에서 값이 다르지만 서로 포함하는 키와 해당 값들: {different_but_contain_values}")
print(f"포함율: {(len(data1)-len(different_values)+len(different_but_contain_values))/len(data1)*100}")
print(f"첫 번째 데이터에만 있는 값: {unique_to_data1}")
print(f"두 번째 데이터에만 있는 값: {unique_to_data2}")
```
